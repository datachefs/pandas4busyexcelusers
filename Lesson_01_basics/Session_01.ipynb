## What is a DataFrame, and How Do You Read A CSV File Into One?

Most of the time in pandas, you're going to be working in what pandas calls a **DataFrame**. It's essentially like an Excel worksheet. When you see the word "DataFrame," just think worksheet.

What is a DataFrame look like? To find out, we're going to work with from the CDC:  Tobacco_Use_1995-2010.csv, which has a breakdown of smokers and non-smokers by state from 1995 to 2010.

But before we do anything in pandas, first we need to tell IPython Notebook to import the pandas library:

```python
import pandas as pd
```

In pandas, when you import a library, most of the time you'll give it a shorter name that makes it a little easier to work with; in this case we gave it "pd", which is the standard abbreviation for pandas.

Okay, now we are ready to read the tobacco file into a DataFrame. The file is a .CSV file, so to read it in and stored in a new DataFrame called tobacco, we write:

```python
tobacco = pd.read_csv("../data/Tobacco_Use_1995-2010.csv")
```

Unlike a lot of programming languages, in pandas we can read in the file with just 1 line of code.

Most of the time when you read in a file, you're going to want to take a look at the first couple of rows to see what the data looks like. To do that, we'll use head

```python
tobacco.head(10)
```

In this case, we asked for the first 10 rows. If we needed to see more of the file, we could just change the number in the parentheses. For example, to read in 20 rows, use:

```python
tobacco.head(20)
```
